```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt 
data_frame_Arrests_under_crime_against_women = pd.read_csv("Arrests_under_crime_against_women.csv")
print("Intial data to know all the attributes and it type without loading all the data")
data_frame_Arrests_under_crime_against_women.head()
```

```python
print(data_frame_Arrests_under_crime_against_women.shape)
print(data_frame_Arrests_under_crime_against_women.info())
print(data_frame_Arrests_under_crime_against_women.describe())
data_frame_Arrests_under_crime_against_women.hist(bins=30)
plt.savefig("image-1.png")
print(plt.show())
from pandas.plotting import scatter_matrix
attributes = ["Year","Persons_Acquitted","Persons_Arrested","Persons_Chargesheeted","Total_Persons_under_Trial","Persons_Trial_Completed"]
scatter_matrix(data_frame_Arrests_under_crime_against_women[attributes])
plt.savefig("image-2.png")
plt.show()
data_frame_Arrests_under_crime_against_women.corr()
```

```python
data_frame_Arrests_under_crime_against_women.drop(labels="Group_Name", axis='columns', inplace=True)
print(data_frame_Arrests_under_crime_against_women.shape)
data_frame_Arrests_under_crime_against_women.head()
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(data_frame_Arrests_under_crime_against_women, test_size=0.25, random_state=94)
print(train_set.shape)
print(train_set.head())
print(test_set.shape)
print(test_set.head())
```

```python
import matplotlib.pyplot as plt
plt.scatter(x = data_frame_Arrests_under_crime_against_women["Persons_Convicted"], y = data_frame_Arrests_under_crime_against_women["Year"],marker="o", c="green")
plt.scatter(x = data_frame_Arrests_under_crime_against_women["Persons_Arrested"], y = data_frame_Arrests_under_crime_against_women["Year"],marker="v", c="red")
plt.xlabel("Persons Convicted vs Persons Arrested")
plt.ylabel("years")
plt.savefig("image.png")
plt.show()
```
