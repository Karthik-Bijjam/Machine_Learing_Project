 ### Clustering

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data_frame_Arrests_under_crime_against_women = pd.read_csv("Arrests_under_crime_against_women.csv")
data_frame_Arrests_under_crime_against_women.describe()
```

```python
import math
from sklearn.cluster import KMeans
X = data_frame_Arrests_under_crime_against_women[["Persons_Acquitted", "Total_Persons_under_Trial"]]
kmeans_classifier = KMeans(n_clusters=2)
kmeans_classifier.fit(X)
print("The distortion: ", kmeans_classifier.inertia_)
RMS_average = math.sqrt(kmeans_classifier.inertia_ / len(X))
print("Averaging and taking the square root gives: ", RMS_average)
print("The centroids: ", kmeans_classifier.cluster_centers_)
print("Assignments: ")
print(kmeans_classifier.labels_)
cluster_x = [Persons_Acquitted for [Persons_Acquitted,Total_Persons_under_Trial] in kmeans_classifier.cluster_centers_]
cluster_y = [Persons_Acquitted for [Persons_Acquitted,Total_Persons_under_Trial] in kmeans_classifier.cluster_centers_]
import matplotlib.pyplot as plt
plt.scatter(data_frame_Arrests_under_crime_against_women['Persons_Acquitted'], data_frame_Arrests_under_crime_against_women['Total_Persons_under_Trial'], c=kmeans_classifier.labels_)
plt.scatter(cluster_x, cluster_y, marker="+", color="red")
plt.xlabel("Persons_Acquitted")
plt.ylabel("Total_Persons_under_Trial")
plt.show()
plt.scatter(data_frame_Arrests_under_crime_against_women['Persons_Acquitted'], data_frame_Arrests_under_crime_against_women['Total_Persons_under_Trial'], c=kmeans_classifier.labels_)
plt.scatter(cluster_x, cluster_y, marker="+", color="red")
plt.xlabel("Persons_Acquitted")
plt.ylabel("Total_Persons_under_Trial")
plt.axis("scaled")
plt.savefig("image.png")
plt.show()
```

```python
import pandas as pd
data_frame_Arrests_under_crime_against_women_copy_1 = data_frame_Arrests_under_crime_against_women.copy()
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(data_frame_Arrests_under_crime_against_women_copy_1, test_size = 0.25, random_state = 123)
data_frame_Arrests_under_crime_against_women_copy_1.describe()
data_frame_Arrests_under_crime_against_women_copy_1.info()
```

```python
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
X_train = train_set[["Persons_Acquitted","Total_Persons_under_Trial","Persons_Arrested"]]
y_train = train_set["Year"]
pca = PCA()
pca.fit(X_train)
cumulative_sums = np.cumsum(pca.explained_variance_ratio_)
d_values = [i+1 for i in range(0,len(cumulative_sums))]
plt.plot(d_values, cumulative_sums)
plt.savefig("image.png")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
X = data_frame_Arrests_under_crime_against_women_copy_1[["Persons_Acquitted","Total_Persons_under_Trial","Persons_Arrested"]]
y = data_frame_Arrests_under_crime_against_women_copy_1["Year"]
from sklearn.metrics import accuracy_score, f1_score
from sklearn.model_selection import KFold
validation_accuracy = []
validation_f1 =[]
fold_and_validate = KFold(n_splits=5, shuffle=True, random_state=145)
for train_set_indices, validation_set_indices in fold_and_validate.split(X):
    cv_train_set = X.iloc[train_set_indices]
    cv_train_target = y.iloc[train_set_indices]
    
    cv_randomForestClassifier = RandomForestClassifier()
    cv_randomForestClassifier.fit(cv_train_set, cv_train_target)
    
    cv_xvalidation = X.iloc[validation_set_indices]
    cv_y_true = y.iloc[validation_set_indices]
    cv_y_predicted = cv_randomForestClassifier.predict(cv_xvalidation)
    
    cv_accuracy_score = accuracy_score(cv_y_true, cv_y_predicted)
    cv_f1_score = f1_score(cv_y_true, cv_y_predicted,  average="weighted")
    validation_accuracy.append(cv_accuracy_score)
    validation_f1.append(cv_f1_score)
    
print("Cross validation accuracies are: ", validation_accuracy)
print("Cross validation f1 scores  are: ", validation_f1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, f1_score
X = data_frame_Arrests_under_crime_against_women_copy_1[["Persons_Acquitted","Total_Persons_under_Trial","Persons_Arrested"]]
y = data_frame_Arrests_under_crime_against_women_copy_1["Year"]
cv_randomForestClassifier = RandomForestClassifier()
cv_randomForestClassifier.fit(X,y)
# Train the classifier
print("Evaluating random forest train set")
X = train_set[["Persons_Acquitted","Total_Persons_under_Trial","Persons_Arrested"]]
y_true = train_set["Year"]
y_predicted = cv_randomForestClassifier.predict(X)
matrix = confusion_matrix(y_true, y_predicted)
print(matrix)
print ("Accuracy is ", accuracy_score(y_true, y_predicted))
print ("F1 is ", f1_score(y_true, y_predicted, average="weighted"))
# Test the classifier
print("Evaluating random forest test set")
X = test_set[["Persons_Acquitted","Total_Persons_under_Trial","Persons_Arrested"]]
y_true = test_set["Year"]
y_predicted = cv_randomForestClassifier.predict(X)
matrix = confusion_matrix(y_true, y_predicted)
print(matrix)
print ("Accuracy is ", accuracy_score(y_true, y_predicted))
print ("F1 is ", f1_score(y_true, y_predicted, average="weighted"))
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix
model = MLPClassifier(
    solver='adam', 
    activation='logistic', 
    alpha=1e-5, 
    hidden_layer_sizes=(200),
    random_state=1,max_iter=20)
X_train = train_set[["Persons_Acquitted","Total_Persons_under_Trial","Persons_Arrested"]]
y_train = train_set["Year"]
model.fit(X_train, y_train)
y_predicted = model.predict(X_train)
matrix = confusion_matrix(y_train, y_predicted)
print("Results for Training set")
print(matrix)
print ("Accuracy is ", accuracy_score(y_train, y_predicted))
# We have to specify how to combine for the multiclassifications
print ("Precision is ", precision_score(y_train, y_predicted, average="weighted"))
print ("Sensitivity is ", recall_score(y_train, y_predicted, average="weighted"))
print ("F1 is ", f1_score(y_train, y_predicted, average="weighted"))
X_test = test_set[["Persons_Acquitted","Total_Persons_under_Trial","Persons_Arrested"]]
y_test = test_set["Year"]
y_predicted = model.predict(X_test)
matrix = confusion_matrix(y_test, y_predicted)
print("\nResults for Test set")
print(matrix)
print ("Accuracy is ", accuracy_score(y_test, y_predicted))
# We have to specify how to combine for the multiclassifications
print ("Precision is ", precision_score(y_test, y_predicted, average="weighted"))
print ("Sensitivity is ", recall_score(y_test, y_predicted, average="weighted"))
print ("F1 is ", f1_score(y_test, y_predicted, average="weighted"))
```
